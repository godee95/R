```python
db = MySQLdb.connect(
'localhost',
    'root',
    '1234',
    'world',
    charset = 'utf8'
)
country_df = pd.read_sql('select * from country',db)
```

```python
sql = '''
select name, continent, population from country where continent = 'asia' or 'europe';
'''

new_country_df = pd.read_sql(sql,db);
new_country_df.head()
```

```python
asia = country_df['Continent'] == 'Asia'
europe = country_df['Continent'] == 'Europe'

new_country_df = country_df[asia | europe][['Name','Continent','Population']]
new_country_df.head()
```
