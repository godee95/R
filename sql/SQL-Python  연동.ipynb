```python
# sakila의 rental 테이블에서 월별 렌탈 수 출력
# 날짜는 오름차순 정렬
# dataframe
# pandas, sql
```

```python
db = MySQLdb.connect(
'localhost',
    'root',
    '1234',
    'sakila',
    charset = 'utf8'
)
```

```python
# sql-1
sql = '''
select month(rental_date) as 월별, count(*) as 렌탈수
	from rental
	group by month(rental_date)
    order by rental_date;
'''
rental_df = pd.read_sql(sql,db)
rental_df
```

```python
# sql-2
sql = '''
select left(rental_date, 7) 'month', rental_id, count(rental_id) 렌탈수
	from rental 
	group by month;
'''
rental_df = pd.read_sql(sql,db)
rental_df.head()
```

```python
# pandas
sql = '''
select * from rental 
'''
rental_df = pd.read_sql(sql,db)
rental_df.head()
```

```python
rental_df = rental_df[['rental_id','rental_date']]
rental_df.head()
```

```python
rental_df['month'] = rental_df['rental_date'].apply(lambda x:str(x)[5:7])
rental_df.head()
```

```python
# rental_id는 unique한 값이므로 이를 count
rental_df.groupby('month')['rental_id'].count()
```

```python
# payment table
# 월별 전체 매출, 평균 매출, 결제 갯수
# 위의 값을 출력하고
# 위의 각각의 값을 dataframe의 칼럼값으로 해서 출력
```

```python
# sql
sql = '''
select month(payment_date) '월', avg(amount) '월 평균매출', 
    count(*) '결제 갯수', sum(amount) '월 매출액'
        from payment
            group by 월;
'''
payment_df = pd.read_sql(sql,db)
payment_df.head()
```

```python
# pandas
sql = '''
select * from payment;
'''
payment_df = pd.read_sql(sql,db)
payment_df.head()
```

```python
payment_df = payment_df[['payment_id','payment_date','amount']]
```

```python
payment_df['month'] = payment_df['payment_date'].apply(lambda x:str(x)[5:7])
payment_df.head()
```

```python
payment_df.groupby('month')['payment_id'].count()
```

```python
# 월별 총 매출액
payment_df.groupby('month')['amount'].sum()
```

```python
payment_df.groupby('month')['amount'].mean()
```

```python
# dataframe형식
# colum별 함수 적용 가능
import numpy as np
payment_df.groupby('month').agg({'amount': [np.mean, np.sum], 'payment_id': np.size})
```
