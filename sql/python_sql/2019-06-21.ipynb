```python
# sakila db
# 고객별 렌탈 횟수가 30회 이상인 유저
# 유저 이름, 유저 이메일 dataframe형태로 출력
# customer, rental
```

```python
db = MySQLdb.connect(
'localhost',
    'root',
    '1234',
    'sakila',
    charset = 'utf8'
)
```

```python
sql = '''
use sakila;
'''
cursor.execute(sql)
```

```python
# sql
sql = '''
select first_name, email, count(rental_id) num_rent
    from customer c join rental r
        on c.customer_id = r.customer_id
            group by c.customer_id
                having num_rent >= 30
                    order by num_rent desc;
'''
result_df = pd.read_sql(sql, db)
result_df.head()
```

```python
# pandas
sql = '''
show tables;
'''
cursor.execute(sql)
```

```python
cursor.fetchall()
```

```python
# pandas
sql = '''
select * from rental;
'''
rental_df = pd.read_sql(sql, db)
rental_df.head(3)
```

```python
sql = '''
select * from customer;
'''
customer_df = pd.read_sql(sql, db)
customer_df.head(3)
```

```python
df = pd.merge(customer_df, rental_df, on = 'customer_id')
df.head(2)
```

```python
df30 = df.groupby('customer_id')['rental_id'].count() >= 30
```

```python
df30[:5]
```

```python
# True인 index만 뽑아서 list에 담음
idx30 = [idx for idx, i in enumerate(df30) if i == True]
```

```python
idx30[:5]
```

```python
# 같은 이름, 같은 customer_id가 너무 많아서 unique한 값을 다시 뽑아내야 함.
df.head()
```

```python
df = df[['first_name', 'email','customer_id']]
```

```python
# unique한 값만 추출해주는 함수 이용.
uniquedf = df.drop_duplicates(['customer_id'], keep = 'last')
uniquedf.head()
```

```python
# drop = True 원래 있던 index 삭제..
uniquedf.set_index('customer_id', drop = True).iloc[idx30]
```

```python
# film 테이블에서 2006년이나 2007년에 출시되었으면서
# 'PG' 등급이거나 'G' 등급의 영화제목
# film table
```

```python
# sql
sql = '''
select title, release_year, rating from film
    where (release_year = 2006 or release_year = 2007) and
        (rating = 'PG' or rating = 'G');
'''
film_df = pd.read_sql(sql, db)
film_df.head(10)
```

```python
# pandas

sql = '''
select * from film;
'''
film_df = pd.read_sql(sql,db)
film_df.head(3)
```

```python
f_2007 = film_df['release_year'] == 2007
f_2006 = film_df['release_year'] == 2006

film_df[f_2006 | f_2007].head(3)
```

```python
f_PG = film_df['rating'] == 'PG' 
f_G = film_df['rating'] == 'G' 

film_df[f_PG | f_G].head(3)
```

```python
film_result = film_df[ (f_2006 | f_2007) & (f_PG | f_G) ]
film_result.head(10)
```

```python
# user별 매출
# customer, payment table
# 내림차순 정렬, customer_id, 매출 뽑기
```

```python
# sql
sql = '''
select distinct c.customer_id ,sum(p.amount) as revenue
    from customer c join payment p
        on c.customer_id = p.customer_id
            group by c.customer_id
                order by revenue;
'''

result_df = pd.read_sql(sql, db)
result_df.head()
```

```python
sql = '''
select
    c.customer_id,
    sum(p.amount) 'revenue'
from
    payment p
    join customer c
    
on c.customer_id = p.customer_id
group by customer_id
order by revenue;
'''
pd.read_sql(sql, db).head()
```
